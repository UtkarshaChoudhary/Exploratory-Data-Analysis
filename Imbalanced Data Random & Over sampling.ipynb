```python
import pandas as pd
```

```python
data=pd.read_csv("Socialnetworkads.csv")
```

```python
data.head(3)
```

```python
data["Purchased"].value_counts()
```

```python
X=data.iloc[:,:-1]
Y=data["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(X_train,Y_train)
```

```python
#47	20000	1
#49	28000	1
#47	30000	1
#29	43000	0
#31	18000	0
```

```python
lr.score(X_test,Y_test)*100
```

```python
lr.predict([[19,19000]])   # This value is correct according to our data 
```

```python
lr.predict([[47,30000]])        #But this value is incorrect acoording to our data because the value_count of 0 is bayes categories
```

### Random Under sampling

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
ru=RandomUnderSampler()
ru_X,ru_Y=ru.fit_resample(X,Y)
```

```python
ru_Y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(ru_X,ru_Y, test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR=LogisticRegression()
LR.fit(X_train,Y_train)
```

```python
LR.score(X_test,Y_test)*100
```

```python
#47	20000	1
#49	28000	1
#47	30000	1
#29	43000	0
#31	18000	0
```

```python
LR.predict([[31,18000]])
```

```python
LR.predict([[49,28000]])
```

## Random Over Sampling

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ro=RandomOverSampler()
ro_X,ro_Y=ro.fit_resample(X,Y)
```

```python
ro_Y.value_counts()
```

```python
X_train1,X_test1, Y_train1, Y_test1= train_test_split(ro_X,ro_Y, test_size=0.2, random_state=42)
```

```python
lr1= LogisticRegression()
lr1.fit(X_train1,Y_train1)
```

```python
lr1.score(X_test1, Y_test1)*100
```

```python
lr1.predict([[49,28000]])
```

```python
lr1.predict([[29,43000]])
```
