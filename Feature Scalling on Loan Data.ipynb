```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("loan.csv")
```

```python
data.head(2)
```

```python
data.isnull().sum()
```

```python
sns.distplot(data["ApplicantIncome"])
plt.show()
```

```python
data.describe()
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
ss.fit(data[["ApplicantIncome"]])
```

```python
data[["ApplicantIncome_ss"]]=pd.DataFrame(ss.transform(data[["ApplicantIncome"]]),columns=["x"])
```

```python
data.head(5)
```

```python
data.describe()
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(data["ApplicantIncome"])

plt.subplot(1,2,2)
plt.title("after")          
sns.distplot(data["ApplicantIncome_ss"])
plt.show()
```

# Normalization

```python
data1=pd.read_csv("loan.csv")
```

```python
data1.head(5)
```

```python
data1.isnull().sum()
```

```python
sns.distplot(data1[["ApplicantIncome"]])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms=MinMaxScaler()
ms.fit(data1[["ApplicantIncome"]])
```

```python
data1[["ApplicantIncome_ms"]]=ms.transform(data1[["ApplicantIncome"]])
```

```python
data1.head(3)
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("before")
sns.distplot(data1[["ApplicantIncome"]])

plt.subplot(1,2,2)
plt.title("after")
sns.distplot(data1[["ApplicantIncome_ms"]])
plt.show()
```
