```python
data=pd.read_csv("Cutlets.csv")
data.head(5)
```

```python
from scipy import stats
```

```python
A=data.iloc[:,0]
B=data.iloc[:,1]
```

```python
A
```

```python
T=stats.ttest_ind(A,B)
T
```
